### 9. We want to use $k$-nearest neighbors to predict whether or not a person is left-handed. What munging/cleaning do we need to do to our $y$ variable in order to explicitly answer this question? Do it.

Answer: 

```python
#There are 3 outcomes for hand; we want to combine 2 and 3.
modified_df["hand"] = modified_df["hand"].apply(lambda x: 1 if (x == 2) or (x==3) else 0)
```

### 11. Let's *(finally)* use $k$-nearest neighbors to predict whether or not a person is left-handed!

> Be sure to create a train/test split with your data!

> Create four separate models, one with $k = 3$, one with $k = 5$, one with $k = 7$, and one with $k = 9$.

> Instantiate and fit your models using GridSearchCV.

```python
columns = [f"Q{i}" for i in range(1, 45)]
X = modified_df[columns]
Y = modified_df["hand"]
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y)
```

```python
k_vals = [3, 5, 7, 9]
weight_possibs = ["uniform", "distance"]
grd = GridSearchCV(KNeighborsClassifier(weights="distance"), param_grid={"n_neighbors": k_vals})
```

```python
grd.fit(X_train, Y_train)
```

```python
grd.best_params_
```

```python
grd.best_score_
```

```python
ser = pd.Series(grd.predict(X_train))
len(ser[ser == 1])
```

```python
recall_score(Y_train, grd.predict(X_train))
```

```python
knn = KNeighborsClassifier(weights=None, n_neighbors = 5)
knn.fit(X_train, Y_train)
```

```python
recall_score(Y_train, knn.predict(X_train))
```

### 21. For your logistic regression models, play around with the regularization hyperparameter, $C$. As you vary $C$, what happens to the fit and coefficients in the model? What do you think this means in the context of this specific problem?

```python
Y_train.describe()
```

```python
from sklearn.metrics import accuracy_score
```

```python
regr = LogisticRegression(C=1000)
regr.fit(X_train, Y_train)
```

```python
print(regr.score(X_train, Y_train))#0.8492670490758445 --> WHY IS THIS ACCURACY STILL THE SAME???
print(regr.score(X_test, Y_test))
print(regr.coef_) # The Coefficients Change
predictions = regr.predict(X_train)
print(predictions)
# It's because the predictions are all 0.
print(accuracy_score(Y_train, predictions))
print(accuracy_score(Y_train, [0 for i in range(len(predictions))]))
```

```python
pred = pd.Series(predictions)
pred.describe()
```

Answer: As I increase C, the coefficients change; but the accuracy remains the same. Upon further analysis, it is because the logistic regression always predicts 0 for the class. This is because the dataset only has 15% 1s. This means that the loss function is minimized by just predicting 0 for everything, giving it an 85% accuracy (especially since there might be almost no correlation with the questions to being left-handed).
