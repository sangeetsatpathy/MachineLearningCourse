```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
xs = 5 + stats.norm.rvs(0, 1, 100_000)
```

```python
xs
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

sns.kdeplot(-np.exp(0.4*xs), color='#e51b24', shade=True, ax=axs[0])
sns.kdeplot(xs, color='#e51b24', shade=True, ax=axs[1])
sns.kdeplot(np.exp(0.4*xs), color='#e51b24', shade=True, ax=axs[2])

axs[0].set_yticks([])
axs[0].set_xticks([])
axs[0].set_title('Left Skew', loc='left')

axs[1].set_yticks([])
axs[1].set_xticks([])
axs[1].set_title('Symmetric', loc='left')

axs[2].set_yticks([])
axs[2].set_xticks([])
axs[2].set_title('Right Skew', loc='left')

plt.tight_layout()
plt.savefig('skew-plot.png', dpi=192)
```

```python
xs = stats.norm.rvs(0, 1, 100_000)
ys = xs + 5
zs = xs[:int(2*100_000/3)] + 11
ws = xs[:int(3*100_000/4)] + 16
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

sns.kdeplot(xs, color='#e51b24', shade=True, ax=axs[0])
sns.kdeplot(np.append(xs, ys), color='#e51b24', shade=True, ax=axs[1])
sns.kdeplot(np.concatenate([xs, ys, zs, ws]), color='#e51b24', shade=True, ax=axs[2])

axs[0].set_yticks([])
axs[0].set_xticks([])
axs[0].set_title('Unimodal', loc='left')

axs[1].set_yticks([])
axs[1].set_xticks([])
axs[1].set_title('Bimodal', loc='left')

axs[2].set_yticks([])
axs[2].set_xticks([])
axs[2].set_title('Multimodal', loc='left')

plt.tight_layout()
plt.savefig('mode-plot.png', dpi=192)
```
